```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("0.5t_convergence.csv", header=0)
df = df.apply(pd.to_numeric, errors='coerce')
df.head()
```

```python
fig, ax = plt.subplots(1,3 , figsize=(15,5))
ax[0].plot(df["MC step"], df["Ratio"])
ax[0].set_xlabel("MC step")
ax[0].set_ylabel("Ratio")
ax[0].grid()

ax[1].plot(df["MC step"], df["Energy"])
ax[1].set_xlabel("MC step")
ax[1].set_ylabel("Energy")
ax[1].grid()

ax[2].plot(df["MC step"], df["Ratio"])
ax[2].set_xlabel("MC step")
ax[2].set_ylabel("Ratio")
ax[2].grid()

fig.tight_layout()
```
