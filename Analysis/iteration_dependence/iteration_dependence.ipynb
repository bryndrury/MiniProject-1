```python
# plot with error bars between the three runs
df["mean_time"] = df[["time (run 1)", "time (run 2)", "time (run 3)"]].mean(axis=1)
df["min_time"] = df[["time (run 1)", "time (run 2)", "time (run 3)"]].min(axis=1)
df["max_time"] = df[["time (run 1)", "time (run 2)", "time (run 3)"]].max(axis=1)

plt.figure(figsize=(8, 6))

plt.errorbar(df["steps"], df["mean_time"], yerr=[df["mean_time"] - df["min_time"], df["max_time"] - df["mean_time"]], 
             color='tab:green', capsize=2, linestyle=" ", label="Mean time with error bars")
plt.scatter(df["steps"], df["mean_time"], 
            color='tab:green', label="Mean time", marker="o", facecolors="white",)

plt.title("20 x 20 Lebwohl-Lasher Varying Steps")
plt.xlabel("Number of steps")
plt.ylabel("Time (s)")

plt.tight_layout()
plt.legend()

plt.savefig("lebwohl_lasher_time.png")
```
